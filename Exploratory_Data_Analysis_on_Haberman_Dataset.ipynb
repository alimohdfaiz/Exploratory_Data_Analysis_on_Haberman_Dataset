PDF

```python
sns.FacetGrid(df, hue = 'status', size =5)\
    .map(sns.distplot,'age')\
    .add_legend()
plt.title("PDF for Age")
plt.xlabel("Age")
plt.ylabel("Density")
plt.show()
```

```python
# for indx, feature in enumerate(list(df.columns)[:-1]):
# for feature in list(df.columns)[:-1]:
#   fg = sns.FacetGrid(df,hue='status',size=5)
#   fg.map(sns.distplot,feature).add_legend()
#   plt.show()

for feature in list(df.columns)[:-1]:
  sns.FacetGrid(df,hue='status',size=5)\
  .map(sns.distplot,feature).add_legend()
  plt.title("PDF for : "+feature)
  plt.show()
```

By seeing the above graphs we can analyze that Age and Operation year does not affecting much on the Survival Status, but nodes is the feature which is affecting the survival status very much.

```python
plt.figure(figsize=(20,5))
for idx, feature in enumerate(list(df.columns)[:-1]):
    
    plt.subplot(1, 3, idx+1)
    print("********* "+feature+" *********")
    
    counts, bin_edges = np.histogram(df[feature], bins=10, density=True)
    print("Bin Edges: {}".format(bin_edges))
    pdf = counts/sum(counts)
    print("PDF: {}".format(pdf))
    cdf = np.cumsum(pdf)
    print("CDF: {}".format(cdf))
    
    plt.plot(bin_edges[1:], pdf,label= "PDF")
    plt.plot(bin_edges[1:], cdf, label="CDF")
    plt.xlabel(feature)
    plt.ylabel("Density")
    plt.legend()
    plt.title("PDF and CDF of Feaure :"+feature)

    # Took Reference From Raghunath D(Medium.com)
```

1. By seeing the PDFs we can analyze that around 20% of people survived for which operation year is 59 to 60 
2. Around 80% of people survived who has nodes less than 5

**BOX PLOT**

```python
fig, axes = plt.subplots(1,3, figsize=(15,5))

for indx,feature in enumerate(list(df.columns)[:-1]):
  sns.boxplot(x = 'status',y=feature,data=df, ax = axes[indx])
  

# sns.boxplot(x = 'status',y='nodes',data=df)
```

1. Above Age Plot reflecting people who survived mostly are from 40 to 60 years age range, and who couldn't survived are between 45 and 60 age groupm
2. Operation Year Plot reflecting people whose operation done between year 60 and 66 are survived mostly and whose operation done between 60 to 65 did not survive.
(There is huge intersection between the above data, so these features are not that useful to find the survival status based on these features)
3. by seeing Nodes plot we can analyse mostly people who survived are having node between 0 to 5, and people who couldn't survive are having nodes between 2 to 5.
(There is less intersection between the data as compared to oher features so this feature is more useful to get the survival status of patients)

**Violin Plots**

```python
fig, axes = plt.subplots(1,3,figsize=(15,5))

for indx, feature in enumerate(list(df.columns)[:-1]):
  sns.violinplot(x='status', y=feature, data=df, ax = axes[indx])
```

1. By seeing Age Violin Plot we can see people having age around 55 are survived more, and people having age around 50 survived less 
2. Year Plot indicates that people whose operation done in the year 65(approx) for those death rate is high.
3. Nodes plot indicates People having nodes 0 to 1 are surviving more and as number increases survival rate is decreasing.
